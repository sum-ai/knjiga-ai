```python
import pandas as pd
data = pd.read_csv('skupovi_podataka/diamonds.csv')
data.head()
```

```python
categoric_columns = ['cut', 'color', 'clarity']

for column in categoric_columns:
  data[column] = data[column].astype('category').cat.codes
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data[['price', 'table', 'carat', 'cut', 'color', 'clarity', 'depth', 'x', 'y', 'z']],
                                                    data['price'], test_size=0.2, random_state=3)
X_train.head()
```

```python
# K-Means klasteriranje - pronalazi grupe sličnih podataka (klastere), ali unaprijed moramo definirat broj klastera.
# Npr. DBSCAN sam pronalazi koliko klastera ima.
# Čitanje: https://towardsdatascience.com/understanding-k-means-clustering-in-machine-learning-6a6e67336aa1
# Dobar primjer: https://static.javatpoint.com/tutorial/machine-learning/images/k-means-clustering-algorithm-in-machine-learning.png

from sklearn.cluster import KMeans
model = KMeans(n_clusters=3) # Sami odredimo koliko bi klastera "moglo" biti.
model.fit(X_train) # Standardno treniramo model na trening podacima

sve_predikcije = model.predict(X_test) # Odredi u koji klaster spadaju uzroci iz testnog skupa
sve_predikcije
```

```python
import numpy as np
trening_predikcije = model.predict(X_train)
sve_predikcije = np.append(sve_predikcije, trening_predikcije)
```
